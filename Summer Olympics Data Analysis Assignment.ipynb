```python
import pandas as pd
df = pd.read_csv("summer.csv") 
```

### 1. In how many cities Summer Olympics is held so far?

```python
len(df['City'].unique())
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df = df[df['Medal'] == 'Gold']

data = []

for sport in df['Sport'].unique():
    data.append([sport, len(df[df['Sport']==sport])])

pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (10,5))
pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending = False).head(5)
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []

for sport in df['Sport'].unique():
    data.append([sport , len(df[df['Sport']  == sport])])

pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).plot(x = 'Sport', y = 'Medal', kind = 'bar', figsize = (10,5))

pd.DataFrame(data,columns = ['Sport','Medal']).sort_values(by='Medal', ascending=False).head(5)
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []

for player in df['Athlete'].unique():
    data.append([player , len(df[df['Athlete']  == player])])

pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head(15).plot(x = 'Athlete', y = 'Medal', kind = 'bar', figsize = (10,5))

pd.DataFrame(data,columns = ['Athlete','Medal']).sort_values(by='Medal', ascending=False).head(5)
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []

for Event in df['Event'].unique():
    data.append([Event , len(df[df['Event']  == Event])])

pd.DataFrame(data,columns = ['Event','Athelete']).sort_values(by='Athelete', ascending=False).head(15).plot(x = 'Event', y = 'Athelete', kind = 'bar', figsize = (10,5))

pd.DataFrame(data,columns = ['Event','Athelete']).sort_values(by='Athelete', ascending=False).head(5)
```

### 8. Which sport is having most female Gold Medalists? (Top 5)
